```python
def getImageUsingPixels(indexNumber):
    imageData=[]
    row=[]
    for i in df.iloc[indexNumber]:
        if len(row)==28:
            imageData.append(row)
            row=[]
        row.append(i+1)
    return imageData
```

```python
allDataset=[]
for i in range(df.shape[0]):
    number = getImageUsingPixels(i)
    allDataset.append(number)
```

```python
for index, row in df.iterrows():
    image = getImageUsingPixels(index)
    data = Image.fromarray(np.array(image).astype(np.uint8))
    if not os.path.isdir(f'dataset/{row["label"]}'): os.mkdir(f'dataset/{row["label"]}')
    data.save(f'dataset/{row["label"]}/{row["label"]}_{index}.png')
#     break
```
